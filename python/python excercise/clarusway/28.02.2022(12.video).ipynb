```python
# WHİLE LOOP DÖNGÜSÜ YENİ SORU
#Kullanıcı bir tane cümle girecek. o cümlenin kelimeleri içindeki kelimelerden en uzun olanını bulacak o kelimenin uzunluğunu 
# LENGTH OF THE LONGEST WORD 
#BU SORUNUN AYRINTILI ANLATIMI İÇİN DEFTERE BAK, CONDİTİON VE İ PARAMETRESİ SONRADAN TANIMLANDI.

sentence = input("Give me a sentenece. :")  #kullanıcıdan bir cümle istiyoruz.
word_list = sentence.split()  #yazılacak o cümleyi list haline getiriyoruz.
longest = 0            #while döngüsü dışında bir değişken tanımladık, uzun kelimeleri biriktirecek bir değişken. bunu ilk etapta 0 atadık yoksa hata verir çalışmaz.
i = 0
while i < len(word_list) :
  if len(word_list) :
    longest = len(word_list[i])
  i += 1
print("The length of the longest word :", longest)
```

```python
numbers = []          #boş bir liste oluşturup, o boş liste içine .append (mutable) ile seçtiğimiz elemanları atmak lazım. bunu for döngüsü ile yapmak.
# append mutable dır listenin orjinalini değiştirir, listeler de mutable dır. her append yaptığımızda list e yeni bir eleman katıyoruz.
for i in range(1, 6) : #bu bir iterable bunu listeye eklemek için önce boş bir liste tanımlamamız lazım. üstteki gibi.
    numbers.append(i)
    print(numbers)
```

```python
word = input("Give me a word :")   #sonunda tre olmaması için if kullanacağız.
counter = 0     # son harften sonra - koymaması için counter tanımladık.
#counter saydırmak için yapılır, teker teker elemanlarını saydırıp fazla gelen - yi atmak için bu değişkeni koyduk, bunu sayaç olarak kullandık.
# sonuncuya - koyma, sadece kendisi olsun.
for i in word :
     counter += 1  # 1<6, 2<6, 3<6, ... 6<6 olunca false verdi döngü bitti.
     if counter < len(word) : #counter kelimenin uzunluğundan küçük olduğu sürece - yazdıracak. 6<6 old. false verip döngüyü bitirecek. (asuman = 6 harf)
       i += "-"               # i nin içinde a var, a nın yanına - nasıl ekleriz? i nin değerini güncellememiz lazım. bu nedenle + ile ekleriz. "a" + "-" = a- oluyor.
     print(i, end = "")  #burada end yazmazsak alt alta sıralar, biz burada end = "" yaparken aşağıya inmesini engelliyoruz.
```
