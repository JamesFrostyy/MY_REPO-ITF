```python
# Kendiniz çözmeden bakmayın hocam
def count_zeros(number): 
    if number == 0:
        return 1
    else:
        count = 0
        while True:
            if number % 10 != 0 or number == 0:
                 break
            count = count + 1 
            number = number / 10

        return count
print(count_zeros(1000))
```

```python
def end_zeros(num: int) -> int:
    # your code here
    text = str(num)
    stripped = text.rstrip('0')
    return len(text) - len(stripped)
```
