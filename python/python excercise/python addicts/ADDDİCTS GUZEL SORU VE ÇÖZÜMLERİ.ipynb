```python
Write a function/functions that checks whether the sentence you get from the user is a palindrome. 
(Do not consider punctuation and special characters. Only consider "alphanumeric" characters.)
input : "ey edip adana'da, pide ye!"

output : "ey edip adana'da, pide ye!" is a palindrome

input : "Was it a car or a cat I saw?"

output : "Was it a car or a cat I saw?" is a palindrome
    
#çözum 1
import string
input_str = 'Was it a car or a cat I saw?'

input_str_final = ''.join(i for i in input_str.replace(' ','').lower() if i not in string.punctuation)

if input_str_final == input_str_final[::-1]:
    print(f'"{input_str}" is a palindrome')
else:
    print(f'"{input_str}" is not a palindrome')

#çözüm 2

import string

def palindrome_control(input_str):    
    input_str_final = ''.join(i for i in input_str.replace(' ','').lower() if i not in string.punctuation)
    
    if input_str_final == input_str_final[::-1]:
        print(f'"{input_str}" is a palindrome')
    else:
        print(f'"{input_str}" is not a palindrome')      

# giris = 'Was it a car or a cat I saw?'
giris = input('Bir metin giriniz: ')

palindrome_control(giris)
```
