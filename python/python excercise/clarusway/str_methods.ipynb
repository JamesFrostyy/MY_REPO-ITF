```python
# Çoklu satır kullanmak:
variable = "MICHAEL"
message = (
    f"Selam {variable}"
    " Nasılsın? "
    f"İsmin doğru değil mi: {variable}"
)
print(message)
# Üçlü-tek/çift-tırnak ile kullanmak:
message = f"""
  {variable}, üc tırnak da böyle kullanılıyor.
"""
print(message)
```

```python
# "Yazının" ilk harfi büyük, diğerleri küçük: str.capitalize()
variable = "MICHAEL JACKSON"
print(f"The popstar is {variable.capitalize()}")
print(f"The popstar is {str.capitalize(variable)}")
```

```python
# "Yazının" tümü küçük harf: str.lower()
variable = "MICHAEL JACKSON"
print(f"the popstar is {variable.lower()}")
print(f"the popstar is {str.lower(variable)}")
```

```python
# "Yazının" tümü büyük harf: str.upper()
variable = "michael jackson"
print(f"THE POPSTAR IS {variable.upper()}")
print(f"THE POPSTAR IS {str.upper(variable)}")
```

```python
# "Kelimelerin" ilk harfi büyük, diğer harfler küçük: str.title()
variable = "micHAEL jackSON"
print(f"The Popstar {variable.title()}")
print(f"The Popstar {str.title(variable)}")
```

```python
# "Her bir harf" için büyük<->küçük: str.swapcase()
variable = "micHAEL jackSON"
print(f"The Popstar {variable.swapcase()}")
print(f"The Popstar {str.swapcase(variable)}")
```

```python
variable = "Bu içinde arama yapılacak yazıdır."
numerator = "0123456789012345678901234567890123"
# Görelim ve kolay sayalım diye geniş yazalım:
print(* variable)
print(* numarator)  
# Yazı başında arama:
print('startswith("B")', variable.startswith("B"))
print('startswith("b")', variable.startswith("b"))
print('startswith("c")', variable.startswith("c"))
print('startswith("a", 10)', variable.startswith("a", 10))
print('startswith("a", 10)', variable.startswith("a", 10, 17))
print('variable.endswith("r")', variable.endswith("r"))
print('variable.endswith(".")', variable.endswith("."))
print('variable.endswith(".", 10)', variable.endswith(".", 10))
print('variable.endswith("y", 10, 17)', variable.endswith("y", 10, 17))
```

```python
# Yazı içinde bulduğunu değiştir: str.replace()
variable = "micha** jacks**"

new_var = variable.replace("**", "--")
# new_var = str.replace(variable, "**", "--")  # Alternatif kullanım.
print(f"The Popstar {new_var}")

new_var = variable.replace("**", "--", 1)  # Değişecek adet tanımlanabilir.
print(f"The Popstar {new_var}")

new_var = variable.replace("a**", "ael").replace("s**", "son")
print(f"The Popstar {new_var}")
```
