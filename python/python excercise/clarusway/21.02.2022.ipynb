```python
a = {'carnation', 'orchid', 'rose', 'violet'}
b = {'rose', 'violet', 'carnation', 'rose', 'orchid', 'rose'}
print(a)
print(b)
```

```python
a.add("lily") #yeni eleman ekledik.
print(a)
```

```python
a.remove('rose') #mevcut bir elemanı sildik.
print(a)
```

```python
a = set('philadelphia')
b = set('dolphin')
print(a)
print(b)
```

```python
#difference metodu
#difference önünde her ne yazarsa onun diğerinden farkı demek istiyor. yani a yazarsak eğer; a nın b den fark.
print(a-b) #1. yol
print(a.difference(b)) #2.yol
```

```python
print(b-a)
print(b.difference(a)) #fark 
```

```python
print(a|b)   # birleşim. burada a birleşim b ya da b birleşi a demek aynı şeydir.
print(a.union(b))
```

```python
#intersection (kesişim) burada a ve b nin yeri değişebilir sorun olmaz. print(b&a) gibi
print(a&b)
print(a.intersection(b))
```

```python
#yeni zellanda başkenti wellington
#usa başkenti washington  BU İKİSİNDEN STRİNG OLARAK KÜMEYE ÇEVİR VE intersection, union, difference metodlarını da kullan.
usa_capt = set('Washington')
nz_capt = set('Wellington')
print(usa_capt) #capital başkent demek.
print(nz_capt)
```

```python
print(usa_capt-nz_capt)    #fark
print(usa_capt.difference(nz_capt))
```

```python
print(usa_capt | nz_capt)             #birleşim
print(usa_capt.union(nz_capt))
```
