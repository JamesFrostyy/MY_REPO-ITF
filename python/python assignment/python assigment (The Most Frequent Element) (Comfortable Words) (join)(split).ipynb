```python
# En sık görülen sayıyı ve sıklığını bulun.
# Bir program yazın;
# Verilen listede en sık görülen numarayı bulur.
# Frekansını hesaplar.
# Aşağıdaki gibi sonucu yazdırır:

numbers = [1, 3, 7, 4, 3, 0, 3, 6, 3]
a = max(numbers, key = numbers.count)
b = numbers.count(max(numbers, key = numbers.count))  # bu satırda   (max(numbers, key = numbers.count) = a yerine direkt a yazabilirdik.)
print("the most frequent number is {} and it was {} times repeated".format(a,b))
```

Rahat bir kelime, her zaman yazdığınız eli değiştirerek yazabileceğiniz bir kelimedir (Q klavyesini kullanarak ve on parmak standardını kullandığınızı varsayarak).
Sözcük her zaman yalnızca a'dan z'ye harflerden oluşan bir dize olacaktır.
Rahat bir kelimeyse True, değilse False döndüren bir program yazın.

```python
left_h = "q, w, e, r, t, a, s, d, f, g, z, x, c, v, b"
right_h = "y, u, i, o, p, h, j, k, l, n, m"
word = set(input("Bir kelime giriniz ").lower())

if word - set(left_h) == set() or word - set(right_h) == set() :
  print(False)
else:
  print(True)
```

```python
right_hand = {"q", "w", "e", "r", "t", "a", "s", "d", "f", "g", "z", "x", "c", "v", "b"}
left_hand = {"y", "u", "i", "o", "p", "h", "j", "k", "l", "n", "m"} 
word = str(input("Lütfen bir kelime girin:"))
x = set(word) 
print(not(x.issubset(right_hand) or x.issubset(left_hand)))
```

```python
# Comfortable Words:
# Left Chars: q, w, e, r, t, a, s, d, f, g, z, x, c, v, b
# Right Chars: y, u, i, o, p, h, j, k, l, n, m
# ---------------------------------------------------------------
chars_left, is_left, chars_right, is_right = 'qwertasdfgzxcvb', False, 'yuiophjklnm', False
check_str = input("input: ")
for c in check_str: # Her bir karakteri sorgulayalım.
  if ((not is_left) and (c in chars_left)): is_left = True # (c in chars_left) yerine -> (chars_left.find(c)>=0) kullanılabilir.
  if ((not is_right) and (c in chars_right)): is_right = True
  # if (is_left and is_right): break # Bulduysan daha fazla arama
print(is_left and is_right)
```

```python
left = ["q", "w", "e", "r", "t", "a", "s", "d", "f", "g", "z", "x", "c", "v", "b"]
right = ["y", "u", "i", "o", "p", "h", "j", "k", "l", "n", "m"]
word = set(input("Enter a word: "))
leftcheck = word.intersection(left)
rightcheck = word.intersection(right)
print(bool(leftcheck) and bool(rightcheck))
```

```python
# JOİN METODU
# Join() dize yöntemi, yinelenebilir bir öğenin (liste, dize, demet) tüm öğelerini bir dize ayırıcıyla ayırarak birleştirerek bir dize döndürür.
# Join den önce (yazdırılan) boşluklar yerine geçer.
# string.join(iterable)

text = ['Python', 'is', 'a', 'fun', 'programming', 'language']

# join elements of text with space

print('***'.join(text))  # burada join öncesi (***) yıldız koyarak bunu daha iyi görebiliriz.
```

```python
# .join() with sets
test = {'2', '1', '3'}
s = ', '
print(s.join(test))

test = {'Python', 'Java', 'Ruby'}
s = '->->'
print(s.join(test))
```

```python
# .join() with dictionaries
test = {'mat': 1, 'that': 2}
s = '->'
# joins the keys only
print(s.join(test))


test1 = {1: 'mat', 2: 'that'}
s = ', '
# this gives error since key isn't string
print(s.join(test1))
```

```python
SPLİT FONKSİYONU
# The syntax of split() is:
# str.split(separator, maxsplit)
split() Parametreler
split() yöntemi en fazla 2 parametre alır:

seperatör (isteğe bağlı)- Bölmelerin gerçekleştiği sınırlayıcı. Sağlanmazsa, dize boşluklara bölünür.
maxsplit (isteğe bağlı) - Maksimum bölme sayısı. Sağlanmadığı takdirde, bölme sayısında bir sınırlama yoktur.

split() Return Value
The split() method returns a list of strings. 
split() yöntemi, dizelerin bir listesini döndürür.
```
