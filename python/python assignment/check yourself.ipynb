```python
print(list.append(5))yaptığınızda bu ekrana sadece None yazdırır arkadaşlar. Neden?
Çünkü her metod/fonsiyon bizim için özel bir iş yapar demiştik. Bazı metod/fonksiyonlar bir sonuç return ederken, append gibi, insert gibi bazı metod/fonksiyonlar ise uygulandığı object'e bir değişiklik yapar ve bırakır, o kadar.
append() fonksiyonu listenin sonuna bir eleman sokar ve liste değişir. Tamam ama herhangi bir değer return etmez. Return etmediği için print() fonksiyonu ile o listeyi göremezsiniz! Siz print() fonksiyonuna append() metodunun yapmış olduğu operasyonu çıktı aldırmaya çalışmış olursunuz. Yani bir eylemi, bir işi çıktı aldırmaya çalışmak gibi bir şeydir bu ve None döndürür. Nedeni dediğimiz gibi append() metodunun bir object return etmemesi, sadece bir operasyon (bir eylem) yapmasıdır.
Aşağıdaki kodları notebook'unuzda hücre hücre  çalıştırıp sonuçları görerek bu meseleyi hemen digest edin arkadaşlar.
hücere:
list((1, 2, 3, 4)) 
# list fonksiyonu ile bir tuple'dan liste yaptık. 
# ve list() fonksiyonunu özelliği gereği bize aynı zamanda bu listeyi return etti.
2. hücre:
print(list((1, 2, 3, 4)))  
# list fonksiyonu bir object return ettiği için, 
# print ile onun çıktısını görebildik.
3. hücre:
list((1, 2, 3, 4)).append(5)
# append() metodu listede değişiklik yaptı. 
# ama herhangi bir object return etmedi.
4. hücre:
print(list((1, 2, 3, 4)).append(6))
# append() metodu herhangi bir şey return etmediği için,
# print ile herhangi bir şey göremedik.
# bu çok normal öyle değil mi.
```
