```python
text = "{:.2f}, {:.3f}, {:.4f}".format(3.1463, 5.367, 7.324324) #burada virgülden sonra kaç basamak alması gerektiği ile ilgili bir kod yazımı yapıldı
print(text)
```

```python
text = "{:.2}, {:.3}, {:.4}".format(3.1463, 5.367, 7.324324) # f leri kaldırınca da her sayıdan olması gerekeni yazıyor. (3.143 burada 2 deyince 3.1 aldı)
print(text)
```

```python
text = "{:.2s}, {:.3s}, {:4s}".format ('a.345', '7.abc', "asu.2022") #en baştan itibaren belirtilen sayı kadar karakter alır.
print(text)
```

```python
text = "{:.2}, {:.3}, {:4}".format ('a.345', '7.abc', "asu.2022") #s kullanmadanda yapılabilir ama buradaki asıl mesele string şeklinde yazılmak zorunda.
print(text)
```

```python
text = "{:>10}".format("test")  # 10 karakter sağa (>10 okun yönü sağı gösteriyor) yaslanacak şekilde yazalım.
print(text)
```

```python
# İNDEXİNG A LİST
colors = ['red', 'blue', 'yellow', 'purple'] #bu şekilde bir değişkene atayarak da yapabiliriz.
print(colors[2])
```

```python
['red', 'blue', 'yellow', 'purple'][2]  #bu şekilde bir değişkene atamadanda yapabiliriz.
```

```python
colors_list = []                       #yeni bir liste tanımladım, boş bir oluşturdum []
colors_list.append(colors)          # sonra bu boş listeye colors listesini ekledim. yukarda görülen boş list in içine colors u ekle,
print(colors_list)      #sonuçta liste içinde liste oldu.
```

```python
#colors_list = [colors] aslında yeni listenin içindeki şey önceki colors listesi, bu nedenle bir elemanı var.
print(len(colors_list))  #burada colors list in içinde sadece colors listesi ([]) var bu nedenle 1 elemanı var.
```

```python
colors_list[1]    #colors list in tek elemanı vardır, bu nedenle hata verir. indexleme 0 dan başlamaktadır. 
```

```python
colors_list[0]  #burada hata vermez çünkü tek elemanlı listenin 0. indexsi direkt listenin kendisini çıkarır.
```
